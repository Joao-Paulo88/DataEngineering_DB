```sql
CREATE OR REPLACE TEMP VIEW vw_vendas_obt
AS
SELECT 
  IdNotaFiscal,
  data_venda,
  mes_venda,
  ano_venda,
  cliente_id,
  cliente_nome,
  marca_carro,
  modelo_carro,
  produto_peca,
  valor_unitario,
  quantidade,
  custo_unitario,
  loja_id,
  loja_nome,
  grupo_loja,
  vendedor_id,
  vendedor_nome,
  faturamento,
  lucro,
  file_path,
  ingest_date
FROM
  tb_fato_vendas fv
INNER JOIN tb_dim_tempo dt 
  ON dt.sk_data_venda = fv.sk_data_venda
INNER JOIN tb_dim_cliente dc 
  ON dc.sk_cliente = fv.sk_cliente
INNER JOIN tb_dim_carro dr 
  ON dr.sk_carro = fv.sk_carro
INNER JOIN tb_dim_produtos dp 
  ON dp.sk_produto = fv.sk_produto
INNER JOIN tb_dim_loja dl 
  ON dl.sk_loja = fv.sk_loja
INNER JOIN tb_dim_vendedor dv 
  ON dv.sk_vendedor = fv.sk_vendedor
ORDER BY 
  IdNotaFiscal;


MERGE INTO tb_vendas_obt tg
USING 
  vw_vendas_obt sc
ON tg.IdNotaFiscal = sc.IdNotaFiscal
WHEN 
  MATCHED 
THEN
  UPDATE SET *
WHEN 
  NOT MATCHED 
THEN
  INSERT *
```
