```sql
USE vendas_pecas.camada_prata;

CREATE OR REPLACE VIEW vw_silver
AS
SELECT DISTINCT id_nf as IdNotaFiscal,
  try_cast(data_venda as date),
  try_cast(cliente_id as bigint),
  cliente_nome,
  marca_carro,
  modelo_carro,
  produto_peca,
  try_cast(valor_unitario as decimal(18,2)),
  try_cast(quantidade as double),
  try_cast(custo_unitario as decimal(18,2)),
  try_cast(loja_id as bigint),
  loja_nome,
  grupo_loja,
  vendedor_id,
  vendedor_nome,
  month(data_venda) AS mes_venda,
  year(data_venda) AS ano_venda,
  TRANSLATE(produto_peca, 'áàãâÁÀÃÂéèêÉÈÊíìîÍÌÎóòõôÓÒÕÔúùûÚÙÛçÇ', 'aaaaAAAAeeeEEEiiiIIIooooOOOOuuuUUUcC') AS produto_peca_sem_acento,
  TRANSLATE(loja_nome, 'áàãâÁÀÃÂéèêÉÈÊíìîÍÌÎóòõôÓÒÕÔúùûÚÙÛçÇ', 'aaaaAAAAeeeEEEiiiIIIooooOOOOuuuUUUcC') AS loja_nome_sem_acento,
  TRANSLATE(grupo_loja, 'áàãâÁÀÃÂéèêÉÈÊíìîÍÌÎóòõôÓÒÕÔúùûÚÙÛçÇ', 'aaaaAAAAeeeEEEiiiIIIooooOOOOuuuUUUcC') AS grupo_loja_sem_acento
FROM vendas_pecas.camada_bronze.tb_bronze
```

```sql
CREATE OR REPLACE TEMP VIEW vw_invalidos
AS
SELECT *
FROM vw_silver VW
WHERE VW.quantidade IS NULL or
VW.quantidade <= 0 or
VW.valor_unitario <= 0 or
VW.custo_unitario <= 0
```

```sql
CREATE OR REPLACE TEMP VIEW vw_validos
AS
SELECT *
FROM vw_silver VW
WHERE VW.quantidade IS NOT NULL and
VW.quantidade > 0 and
VW.valor_unitario > 0 and
VW.custo_unitario > 0
```

```sql
USE vendas_pecas.camada_prata;

CREATE OR REPLACE TABLE tb_quarentena
AS
SELECT *
FROM vw_invalidos
```

```sql
USE vendas_pecas.camada_prata;

CREATE OR REPLACE TABLE tb_prata
AS
SELECT *
FROM vw_validos
```
